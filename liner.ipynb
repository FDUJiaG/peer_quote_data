```python
import pandas as pd
```

```python
df0 = pd.read_excel("基金新进重仓.xlsx")
```

```python
df0
```

```python
col_name = ["code", "name", "s1", "s2", "s3", "s4", "s5", "s6", "s7", "s8", "s9", "s10"]
```

```python
df_out = pd.DataFrame(columns=col_name)
df_out
```

```python
df_values = df0.values
```

```python
lst = df_values[0]
lst[0]
```

```python
lows = df0.shape[0]
```

```python
count_list = list()

for idx in range(lows):
    lst = df_values[idx]
    lst_set = set(lst[2:12])
    lst_dct = dict()
    lst_dct[col_name[0]] = lst[0]
    lst_dct[col_name[1]] = lst[1]
    
    if lst[2] != 0 and lst[-10] != 0:
        sdx = 2
        for jdx in range(-10, 0):
            if lst[jdx] not in lst_set:
                lst_dct[col_name[sdx]] = lst[jdx]
                count_list.append(lst[jdx])
                sdx += 1
            
    df_out = df_out.append(lst_dct, ignore_index=True)

print(df_out)
```

```python
df_count = pd.DataFrame(count_list, columns=["stock"])
```

```python
df_count.value_counts().to_excel("counts.xlsx")
```
