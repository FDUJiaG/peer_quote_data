```python
def get_time(date=False, utc=False, msl=3):
    if date:
        time_fmt = "%Y-%m-%d %H:%M:%S.%f"
    else:
        time_fmt = "%H:%M:%S.%f"

    if utc:
        return datetime.utcnow().strftime(time_fmt)[:(msl-6)]
    else:
        return datetime.now().strftime(time_fmt)[:(msl-6)]


def print_info(status="I"):
    return "\033[0;33;1m[{} {}]\033[0m".format(status, get_time())
```

```python
def get_note(df, short_title, tzz_name):
    df_temp = df[df["投资者名称"] == tzz_name]
    note_dict = dict(df_temp["备注"].value_counts())
    note_len = len(note_dict)
    
    note = ""
    desc_note = ""
    if note_len == 1:
        note_key = list(note_dict.keys())[0]
        if "有效" not in note_key:
            note = note_key
            if "无效" in note:
                desc_note += short_title + str(note_dict[note_key]) + "个" + note[:2]
            else:
                desc_note += short_title + str(note_dict[note_key]) + "个" + note[::2]
    elif note_len == 2:
        note_key = list(note_dict.keys())
        note_sum = sum(note_dict.values())
        
        for note_item in note_key:
            if "有效" in note_item:
                note_key.remove(note_item)
        if len(note_key) == 1:
            note += note_key[0]
            if "无效" in note:
                desc_note += short_title + str(note_sum) + "个配售对象中" + str(note_dict[note]) + "个" + note[:2]
            else:
                desc_note += short_title + str(note_sum) + "个配售对象中" + str(note_dict[note]) + "个" + note[::2]
        else:
            for note_item in note_key:
                note += note_item
                desc_note += short_title
                if "无效" in note_item:
                    desc_note += str(note_dict[note_item]) + "个" + note[:2]
                else:
                    desc_note += str(note_dict[note_item]) + "个" + note[::2]
    else:
        note = "其他情况"

    return note, desc_note

short_title = "万家"
tzz_n = "万家基金管理有限公司"
get_note(df, short_title, tzz_n)
```

```python
def getChars(length):
    return [getChar(index) for index in range(length)]

def getChar(number):
    factor, moder = divmod(number, 26)  # 26 字母个数
    modChar = chr(moder + 65)           # 65 -> 'A'
    if factor != 0:
        modChar = getChar(factor-1) + modChar # factor - 1 : 商为有效值时起始数为 1 而余数是 0
    return modChar
```
